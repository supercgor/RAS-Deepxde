```python
# burger
import numpy as np
import os
import sys
from utils.funcspace import COS, GRF
from dataset import burger_solver
import matplotlib.pyplot as plt
#funcspace = COS(N=2, omega="2 * np.pi")
funcspace = GRF(length_scale=1.0, kernel="ExpSineSquared")
feature = funcspace.random(20)
x = np.linspace(0, 1, 99)
y = funcspace.eval_batch(feature, x)
print(y.shape)
fig, axs = plt.subplots(4, 5, figsize=(20, 16))
for i, ax in enumerate(axs.flatten()):
    ax.plot(x, y[i], label="cos")
# Title
plt.suptitle("Cosine Basis Functions, $N=20$\n", fontsize=20)
plt.tight_layout()
plt.show()

u, xt = burger_solver(y)
u = u.reshape(u.shape[0], -1)

fig, axs = plt.subplots(4, 5, figsize=(20, 16))
for i, ax in enumerate(axs.flatten()):
    ax.scatter(xt[...,0], xt[...,1], c=u[i], cmap="jet")
plt.show()
```

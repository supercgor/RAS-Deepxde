```python
import numpy as np
import os
import sys
from utils.funcspace import COS
import matplotlib.pyplot as plt
cos = COS(N=5, omega="2 * np.pi")
feature = cos.random(20)
x = np.linspace(0, 1, 99)
y = cos.eval_batch(feature, x)
print(y.shape)
fig, axs = plt.subplots(4, 5, figsize=(20, 16))
for i, ax in enumerate(axs.flatten()):
    ax.plot(x, y[i], label="cos")
# Title
plt.suptitle("Cosine Basis Functions, $N=20$\n", fontsize=20)
plt.tight_layout()
plt.show()
```
